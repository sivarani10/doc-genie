```python
import ast
import gradio as gr
import autopep8


class DocGenieAnalyzer:

    @staticmethod
    def extract_function_signature(code):
        tree = ast.parse(code)
        for node in tree.body:
            if isinstance(node, ast.FunctionDef):
                params = []
                for arg in node.args.args:
                    params.append({
                        "name": arg.arg
                    })

                return {
                    "name": node.name,
                    "params": params
                }
        return None


    @staticmethod
    def generate_google_docstring(signature):

        func_name = signature["name"]

        docstring = '"""\n'
        docstring += f"Executes {func_name} operation with conditional logic performing mathematical calculoation and returns the computed result .\n\n"

        docstring += "Args:\n"
        for param in signature["params"]:
            docstring += f"    {param['name']} (Any): Parameter for controlling {param['name']} behavior.\n"

        docstring += "\nReturns:\n"
        docstring += "    Any: The result of the computation.\n"

        docstring += '"""'

        return docstring


def process_code(code):
    signature = DocGenieAnalyzer.extract_function_signature(code)

    if not signature:
        return "No function detected."

    docstring = DocGenieAnalyzer.generate_google_docstring(signature)

    formatted_code = autopep8.fix_code(code)
    lines = formatted_code.split("\n")

    # Insert docstring after function definition
    lines.insert(1, "    " + docstring.replace("\n", "\n    "))

    return "\n".join(lines)


with gr.Blocks() as demo:
    gr.Markdown("# рҹ“ҳ Doc-Genie - Python Docstring Generator")

    code_input = gr.Code(label="Enter Python Function", language="python")

    generate_btn = gr.Button("Generate Docstring")

    output_code = gr.Code(label="Function with Docstring", language="python")

    generate_btn.click(process_code, inputs=code_input, outputs=output_code)

demo.launch()
```
